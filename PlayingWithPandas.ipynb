```python
# JOINS 

# - Left  : uses keys only from the left frame (SQL : Left outer join)
# - Right : uses keys only from the right frame (SQL : Right outer join)
# - Outer : uses the union of keys from both the frames (SQL : Full outer join) 
# - Inner : uses intersection of keys from both the frames (SQL : Inner joins) 
```

```python
left_frame = pd.DataFrame({'key' : range(5),   'left_value':  ['a', 'b', 'c', 'd', 'e']})
right_frame = pd.DataFrame({'key': range(2,7), 'right_value': ['f', 'g', 'h', 'i', 'j']})

print(left_frame.set_index('key').head());
print('\n')
print(right_frame.set_index('key').head());
```

```python
pd.merge(left_frame, right_frame, on='key', how='inner')
```

```python
pd.merge(left_frame, right_frame, on='key', how='left')
```

```python
pd.merge(left_frame, right_frame, on='key',how='right')
```

```python
pd.merge(left_frame, right_frame, on='key', how='outer')
```
